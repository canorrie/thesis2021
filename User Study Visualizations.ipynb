```python
def calculate_score(l):
  total = 0
  for i in l:
    counter = 1
    temp = 0
    for x in i:
      temp += (counter*x)/6
      counter +=1
    total += temp
  return total/len(l)
print(calculate_score([lq01,lq04]))
print(calculate_score([lq05,lq08]))
print(calculate_score([lq09,lq12,lq14]))
print(calculate_score([lq15,lq18,lq20]))
```
